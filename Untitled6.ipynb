```python
from sklearn.naive_bayes import GaussianNB

data = {
    'Symptoms': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes'],
    'Age': [35, 45, 50, 28, 55, 38, 42, 60, 30, 48, 25, 32, 40, 50, 35, 52, 29, 46, 37, 43],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Drug Type': ['Heroin', 'Cocaine', 'Heroin', 'Cocaine', 'Heroin', 'Cocaine', 'Cocaine', 'Heroin', 'Heroin', 'Cocaine', 'Heroin', 'Cocaine', 'Cocaine', 'Heroin', 'Heroin', 'Heroin', 'Cocaine', 'Cocaine', 'Heroin', 'Cocaine'],
    'Blood Alcohol Level': ['High', 'Normal', 'High', 'Normal', 'High', 'Normal', 'High', 'Normal', 'High', 'Normal', 'High', 'Normal', 'Normal', 'High', 'Normal', 'High', 'Normal', 'High', 'Normal', 'High'],
    'Mental Health Condition': ['No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No'],
    'Drug Overdose': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes']
}

symptoms_mapping = {'Yes': 1, 'No': 0}
gender_mapping = {'Male': 1, 'Female': 0}
drug_type_mapping = {'Heroin': 1, 'Cocaine': 0}
bal_mapping = {'High': 1, 'Normal': 0}
mh_condition_mapping = {'Yes': 1, 'No': 0}
overdose_mapping = {'Yes': 1, 'No': 0}

data['Symptoms'] = [symptoms_mapping[val] for val in data['Symptoms']]
data['Gender'] = [gender_mapping[val] for val in data['Gender']]
data['Drug Type'] = [drug_type_mapping[val] for val in data['Drug Type']]
data['Blood Alcohol Level'] = [bal_mapping[val] for val in data['Blood Alcohol Level']]
data['Mental Health Condition'] = [mh_condition_mapping[val] for val in data['Mental Health Condition']]
data['Drug Overdose'] = [overdose_mapping[val] for val in data['Drug Overdose']]

X = [data['Symptoms'], data['Age'], data['Gender'], data['Drug Type'], data['Blood Alcohol Level'], data['Mental Health Condition']]
X = list(zip(*X))
y = data['Drug Overdose']

clf = GaussianNB()
clf.fit(X, y)

input_data = [
    [1, 40, 1, 1, 1, 0],
    [0, 28, 0, 0, 0, 1],
    [1, 55, 1, 1, 1, 1],
    [0, 42, 0, 0, 1, 1],
    [1, 30, 1, 1, 0, 0],
    [0, 48, 0, 0, 0, 1],
    [1, 25, 1, 1, 1, 1],
    [0, 32, 0, 0, 0, 0],
    [1, 40, 1, 1, 1, 0],
    [0, 50, 0, 0, 1, 1]
]

predictions = clf.predict(input_data)
for i, prediction in enumerate(predictions):
    print(f"Predicted Drug Overdose death with input {i+1}: {'Yes' if prediction == 1 else 'No'}")
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd

data = {
    'User_ID': [1, 2, 3, 4, 5],
    'Drugs': [['Cocaine', 'Marijuana'],
              ['Marijuana', 'LSD', 'Ecstasy'],
              ['Cocaine', 'Heroin'],
              ['Marijuana', 'Ecstasy'],
              ['Cocaine', 'LSD']]
}

df = pd.DataFrame(data)

df['Drugs'] = df['Drugs'].apply(lambda x: ','.join(x))

df_encoded = df['Drugs'].str.get_dummies(sep=',')

frequent_itemsets = apriori(df_encoded, min_support=0.2, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(rules)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'User_ID': [1, 2, 3, 4, 5],
    'Drugs': [['Cocaine', 'Marijuana'],
              ['Marijuana', 'LSD', 'Ecstasy'],
              ['Cocaine', 'Heroin'],
              ['Marijuana', 'Ecstasy'],
              ['Cocaine', 'LSD']],
    'Heavy_User': [1, 0, 1, 0, 1]
}

df = pd.DataFrame(data)

df['Drugs'] = df['Drugs'].apply(lambda x: ','.join(x))

df_encoded = df['Drugs'].str.get_dummies(sep=',')

X = df_encoded
y = df['Heavy_User']

clf = DecisionTreeClassifier(random_state=42)

clf.fit(X, y)

plt.figure(figsize=(20, 10))
plot_tree(clf, filled=True, feature_names=df_encoded.columns, class_names=['Non-Heavy User', 'Heavy User'])
plt.show()
```
